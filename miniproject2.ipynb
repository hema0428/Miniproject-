```python
newdataset = SD.drop(['Region', 'Country', 'Item_Type', 'Sales_Channel', 'Order_Priority', 'Order_Date', 'Year', 'Order_ID', 'Ship_Date', 'Units_Sold','Total profit','unit_per Profit'], axis=1)
```

```python
x=newdataset.iloc[:,1].values.reshape(-1,1)
y=newdataset.iloc[:,0].values.reshape(-1,1)
```

```python
pt.scatter(x,y)
pt.xlabel('MAKING COST')
pt.ylabel('SELLING PRICE')
pt.title('LINEAR REGRESSION')
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
Y_predict=lm.predict(X_test)
```

```python
Y_predict
```

```python
SALE=pd.DataFrame(X_test,columns=['Unit_MakingCost'])
```

```python
SALE
```

```python
SALE['AVAILDATA']=Y_test
```

```python
SALE
```

```python
pt.scatter(x,y,color='blue')
pt.plot(X_test,Y_predict,color='pink')
pt.xlabel('UNIT_MAKINGCOST')
pt.ylabel('UNIT_SELLINGPRICE')
pt.title('linear regression')
pt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(Y_test,Y_predict)
```

```python
mean_absolute_error(Y_test,Y_predict)
```

```python
r2_score(Y_test,Y_predict)
```
